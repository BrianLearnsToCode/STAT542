```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp

from sklearn.model_selection import GridSearchCV

from skfda.misc.hat_matrix import NadarayaWatsonHatMatrix
from skfda.ml.regression._kernel_regression import KernelRegression

from skmisc.loess import loess
```

```python
url = "https://liangfgithub.github.io/Data/Coding3_Data.csv"
df = pd.read_csv(url)
```

```python
x = df['x'].to_numpy()
y = df['y'].to_numpy()
n = len(x)
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
```

```python
loocv = []
gcv = []
myspan = [0.05 * i + 0.2 for i in range(15)]
for sp in myspan:
  model = loess(x, y, span=sp)
  yhat = model.predict(x).values
  S = model.outputs.diagonal
  looerr = sum(((y - yhat)/(np.ones(n) - S))**2)/n
  gcverr = sum(((y - yhat)/((1-sum(S)/n)*np.ones(n)))**2)/n
  loocv.append(looerr)
  gcv.append(gcverr)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].scatter(myspan,loocv)
axes[0].set_xlabel("span")
axes[0].set_ylabel("LOOCV error")
axes[1].scatter(myspan,gcv)
axes[1].set_xlabel("span")
axes[1].set_ylabel("GCV error")

fig.set_figwidth(10)
fig.tight_layout()
```

```python
best_span1 = myspan[np.argmin(loocv)]
best_span2 = myspan[np.argmin(gcv)]
print('best span for LOOCV is %f'%best_span1)
print('best span for GCV is %f'%best_span2)
```

```python
model = loess(x, y, span=0.5)
x_new = np.linspace(x[0],x[-1],100)
y_true = np.sin(12*(x_new+0.2))/(x_new + 0.2)
y_fit = model.predict(x_new).values
```

```python
plt.scatter(x,y)
plt.plot(x_new,y_true)
plt.plot(x_new,y_fit,'--')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(["data", "true curve", "fitted curve"], loc="upper right")

plt.tight_layout()
```

Part II

```python
from scipy.interpolate import splev, interp1d
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Sales_Transactions_Dataset_Weekly.csv')
df = df.loc[:,:'W51']
df_mean = df.loc[:,'W0':].mean(axis=1)
columns = ['W' + str(i) for i in range(52)]
for column in columns:
  df[column] = df[column] - df_mean
X = df.loc[:,'W0':].to_numpy()
X.shape
```

```python
def ns(x, df=None, knots=None, boundary_knots=None, include_intercept=False):
    degree = 3

    if boundary_knots is None:
        boundary_knots = [np.min(x), np.max(x)]
    else:
        boundary_knots = np.sort(boundary_knots).tolist()

    oleft = x < boundary_knots[0]
    oright = x > boundary_knots[1]
    outside = oleft | oright
    inside = ~outside

    if df is not None:
        nIknots = df - 1 - include_intercept
        if nIknots < 0:
            nIknots = 0

        if nIknots > 0:
            knots = np.linspace(0, 1, num=nIknots + 2)[1:-1]
            knots = np.quantile(x[~outside], knots)

    Aknots = np.sort(np.concatenate((boundary_knots * 4, knots)))
    n_bases = len(Aknots) - (degree + 1)

    if any(outside):
        basis = np.empty((x.shape[0], n_bases), dtype=float)
        e = 1 / 4 # in theory anything in (0, 1); was (implicitly) 0 in R <= 3.2.2

        if any(oleft):
            k_pivot = boundary_knots[0]
            xl = x[oleft] - k_pivot
            xl = np.c_[np.ones(xl.shape[0]), xl]

            # equivalent to splineDesign(Aknots, rep(k.pivot, ord), ord, derivs)
            tt = np.empty((xl.shape[1], n_bases), dtype=float)
            for j in range(xl.shape[1]):
                for i in range(n_bases):
                    coefs = np.zeros((n_bases,))
                    coefs[i] = 1
                    tt[j, i] = splev(k_pivot, (Aknots, coefs, degree), der=j)

            basis[oleft, :] = xl @ tt

        if any(oright):
            k_pivot = boundary_knots[1]
            xr = x[oright] - k_pivot
            xr = np.c_[np.ones(xr.shape[0]), xr]

            tt = np.empty((xr.shape[1], n_bases), dtype=float)
            for j in range(xr.shape[1]):
                for i in range(n_bases):
                    coefs = np.zeros((n_bases,))
                    coefs[i] = 1
                    tt[j, i] = splev(k_pivot, (Aknots, coefs, degree), der=j)

            basis[oright, :] = xr @ tt

        if any(inside):
            xi = x[inside]
            tt = np.empty((len(xi), n_bases), dtype=float)
            for i in range(n_bases):
                coefs = np.zeros((n_bases,))
                coefs[i] = 1
                tt[:, i] = splev(xi, (Aknots, coefs, degree))

            basis[inside, :] = tt
    else:
        basis = np.empty((x.shape[0], n_bases), dtype=float)
        for i in range(n_bases):
            coefs = np.zeros((n_bases,))
            coefs[i] = 1
            basis[:, i] = splev(x, (Aknots, coefs, degree))

    const = np.empty((2, n_bases), dtype=float)
    for i in range(n_bases):
        coefs = np.zeros((n_bases,))
        coefs[i] = 1
        const[:, i] = splev(boundary_knots, (Aknots, coefs, degree), der=2)

    if include_intercept is False:
        basis = basis[:, 1:]
        const = const[:, 1:]

    qr_const = np.linalg.qr(const.T, mode='complete')[0]
    basis = (qr_const.T @ basis.T).T[:, 2:]

    return basis
```

```python
weeks = np.array([i for i in range(52)])

F = ns(weeks, df = 9)
F_mean = F.mean(axis=0)
F = F - F_mean

B = (np.linalg.inv(F.T @ F) @ F.T @ X.T).T
```

```python
n_clusters = 6
kmeans = KMeans(n_clusters=n_clusters, random_state = 4457)
kmeans.fit(B)

labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_ @ F.T
```

Use reduced features to group time series, the average time series for each group is relatively smooth

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.subplots_adjust(hspace=0.5)

for cluster_num in range(n_clusters):
    row = cluster_num // 3  # Determine the row for the subplot
    col = cluster_num % 3   # Determine the column for the subplot

    # Filter time series data based on cluster label
    cluster_data = X[labels == cluster_num]

    # Plot each time series in the cluster
    for series in cluster_data:
        axs[row, col].plot(series, color='gray', alpha=0.5)  # Adjust color and alpha as needed

    # Plot the cluster center
    axs[row, col].plot(cluster_centers[cluster_num], color='r', linewidth=2)

    axs[row, col].set_xlabel('Weeks')
    axs[row, col].set_ylabel('Weekly Sales')
```

Part III

```python
from sklearn.decomposition import PCA
import time
```

```python
def mypca(X_train, X_test, y_train, y_test, eps=1e-10):
  n_train = len(y_train)
  n_test = len(y_test)
  pca = PCA()


  X_train = X_train - X_train.mean(axis = 0)
  pca.fit(X_train)
  X_train = pca.transform(X_train)
  eigs = pca.explained_variance_
  cutoff = len(eigs[eigs > eps])
  X_train = X_train[:,:cutoff]
  coefs = y_train.T @ X_train / np.sum(X_train**2, axis=0)
  b0 = np.mean(y_train)

  y_train_pred = b0 + X_train @ coefs.T
  train_err = np.sum((y_train - y_train_pred)**2)/n_train

  X_test = X_test - X_test.mean(axis = 0)
  X_test = pca.transform(X_test)
  X_test = X_test[:,:cutoff]
  y_test_pred = b0 + X_test @ coefs.T
  test_err = np.sum((y_test-y_test_pred)**2)/n_test

  return np.log(train_err), np.log(test_err)
```

```python
url = 'https://liangfgithub.github.io/F22/Coding3_dataH.csv'
df = pd.read_csv(url, header = None)

X = df.iloc[:,1:]
y = df[0]

X.shape, y.shape
```

```python
np.random.seed(4457)

epochs = 30
n = len(y)
cols = [i for i in range(5,241)]
m = len(cols)
test_errs = [[] for _ in range(m)]

start = time.time()
for epoch in range(epochs):
  indices = np.arange(0,n)
  np.random.shuffle(indices)
  test_ind = indices[:int(np.floor(0.75*n))]
  train_ind = indices[len(test_ind):]

  for i in range(m):
    col = cols[i]
    X_new = X.iloc[:,:col]

    X_train = X_new.iloc[train_ind]
    y_train = y[train_ind]
    X_test = X_new.iloc[test_ind]
    y_test = y[test_ind]

    train_err, test_err = mypca(X_train, X_test, y_train, y_test)
    test_errs[i].append(test_err)
  end = time.time()
  ctime = end - start
  print('finish epoch %i, %0.2f seconds eclapsed'%(epoch,ctime))

errs = np.median(np.array(test_errs),axis=1)
```
